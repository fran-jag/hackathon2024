```python
#Get the parameters of the mutants to make 0 to simulate KOs
mutants = {'traf2':['cIAP1_2____TRAF2_k1'],
            'traf5': ['cIAP1_2____TRAF5_k1'],
            'traf2_5': ['cIAP1_2____TRAF2_k1', 'cIAP1_2____TRAF5_k1'],
            'traf6': ['TRADD____TRAF6_k1'],
            'rip1': ['TRAF2____RIP1_k1', 'TRAF5____RIP1_k1', 'TRAF6____RIP1_k1'],
            'tradd': ['TNFR1____TRADD_k1']
             }
```

```python
#Max value of WT as normalization
norm = 0.044176444402908084

#Loop over mutant dict to turn each parameter to 0
for key, value_list in mutants.items():
    r.resetAll() #Make sure to reset all before every loop
    for param in value_list:
        setattr(r, param, 0)
    sim = r.simulate(0, 1800, selections=['time','IkBa'])
    plt.plot([x[0]/60 for x in sim], [x[1]/norm for x in sim], '--')    

#Simulate WT
r.resetAll()
sim = r.simulate(0, 1800, selections=['time','IkBa'])
plt.plot([x[0]/60 for x in sim], [x[1]/norm for x in sim], c='black')

#Setup plot legends, labels, etc
plt.legend(['TRAF2 KO', 'TRAF5 KO', 'TRAF2/5 DKO', 'TRAF6 KO', 'RIP1 KO', 'TRADD KO', 'WT'])
plt.xlabel('time (min)')
plt.ylabel('IkBa activation (a.u)')
plt.ylim([0, 2])
plt.show()
```
