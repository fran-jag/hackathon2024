```python
import matplotlib.pyplot as plt
#plot WT as test
fig, axs = plt.subplots(dpi=100)
time = [x[0] for x in ikba_test]
ikba_conc = [x[1] for x in ikba_test]

axs.plot(time, ikba_conc, 'r--');
```

```python
#Parameters modified for mutants from Model B 
mutants = {'traf2':['reaction_48'],
            'traf5': ['reaction_4'],
            'traf2_5': ['reaction_48', 'reaction_4'],
            'traf6': ['reaction_3'],
            'rip1': ['reaction_5', 'reaction_6', 'reaction_7'],
            'tradd': ['reaction_1']
             }
```

```python
#Simulate WT
r.resetAll()
sim = r.simulate(0, 1800, selections=['time','species_53'])

#normalization factor as the max value of WT
norm = max([x[1] for x in sim])

#Plot WT
plt.plot([x[0]/60 for x in sim], [x[1]/norm for x in sim], c='black')

#Plot mutants
for key, value_list in mutants.items():
    r.resetAll()
    for param in value_list:
        setattr(r, param+'_k1', 0)
    sim = r.simulate(0, 1800, selections=['time','species_53'])
    plt.plot([x[0]/60 for x in sim], [x[1]/norm for x in sim], '--')    


#Setup plot
plt.legend(['WTF', 'TRAF2 KO', 'TRAF5 KO', 'TRAF2/5 DKO', 'TRAF6 KO', 'RIP1 KO', 'TRADD KO'], loc='upper left')
plt.xlabel('time (min)')
plt.ylabel('IkBa activation (a.u)')
plt.ylim([0, 2])
plt.show()
```
